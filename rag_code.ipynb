```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

from langchain_community.document_loaders import PyPDFLoader
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=20,
    length_function=len
)
```

```python
# Absolute file path to your document
pdf_path = r"C:\Users\USER\Desktop\tensorvirtual\rag-gemini\data\Nigeria_Data_Protection_Act2023.pdf"
```

```python
# 1. Load the PDF
print("Loading PDF document...")
loader = PyPDFLoader(pdf_path)
documents = loader.load()
chunks =  text_splitter.split_documents(documents)

print(f"Total created chunks: {len(chunks)}")
```

```python
import tiktoken

# 1. Get the exact character count from your LangChain documents
# (Assuming 'documents' is the list of chunks generated by your splitter)
total_characters = sum(len(doc.page_content) for doc in documents)
print(f"Total characters in documents: {total_characters}")

```

```python
# 2. Estimate the number of tokens (assuming GPT-3.5 Turbo tokenizer)
encoding = tiktoken.encoding_for_model("gpt-3.5-turbo")
total_tokens = sum(len(encoding.encode(doc.page_content)) for doc in documents)
print(f"Total tokens in documents: {total_tokens}")
```
